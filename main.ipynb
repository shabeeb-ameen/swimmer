```python
run_directory="run2/"

def draw_graph(run_directory='run/', textfile='0.0.txt'):
    if not os.path.exists(run_directory+"graphs/"):
        os.mkdir(run_directory+"graphs/")
    G = nx.Graph()
        
    with open(run_directory+textfile) as f:

        lines = f.readlines()
        time = float(lines[0].split()[1])
        nodesFlag = False
        edgeFlag = False
        for line in lines:
            if line.split()[0] == "Nodes:":
                nodesFlag = True
                continue
            if line.split()[0] == "Edges:":
                nodesFlag = False
                edgeFlag = True
                continue
            if nodesFlag:
                id = int(line.split()[0])
                coord=(float(line.split()[1]),float(line.split()[2]))
                G.add_node(id, pos=coord)
            if edgeFlag:
                id = int(line.split()[0])
                G.add_edge(int(line.split()[1]),int(line.split()[2]))
    pos = nx.get_node_attributes(G, 'pos')

# draw the graph
    nx.draw(G, pos, with_labels=True)

# show the graph
    lim=2

    plt.xlim(-lim,lim)
    plt.ylim(-lim,lim)
    plt.text(0.9, 0.9, 'Time {}'.format(round(time,3)), fontsize=12, color='red', ha='right', va='top', transform=plt.gca().transAxes)

    plt.savefig(run_directory+"graphs/{}.png".format(textfile.split('.')[0]))
    plt.close()
    return


file_names = []
# Get all files in the folder
for filename in os.listdir(run_directory):
    # Check if the file is a text file
    if filename.endswith(".txt"):
        file_names.append(filename)

for file in file_names:
    draw_graph(run_directory,file)


images = []

for filename in sorted(glob.glob(run_directory+'graphs/*.png')): #assuming png
    images.append(Image.open(filename))

# Save the images as a gif
images[0].save(run_directory+'animated.gif', save_all=True, append_images=images[1:], optimize=False, duration=500, loop=1)
```
